## Project Specific Imports

```python
from src.error_analysis import run_error_analysis_inlier_noise
from src.plotting import plot_error_inlier_noise
```

## Problem Set-Up #1

```python
%%capture
# Dimensions
ambient_dim = 200
intrinsic_dim = 2

# Number of outliers and inliers
outlier_points = [100]
inlier_points = [20]

# Spreads
inlier_spreads = [0.5]
outlier_spreads = [np.float(intrinsic_dim)/np.float(ambient_dim) * 0.5]
inlier_noise_spreads = [0.001, 0.005, 0.01, 0.025, 0.05, 0.1, 0.25,0.35, 0.5, 1.0] # Gaussian noise variance

# repititions
repititions = 20

# Sampling
sampling = "uniform"
method = 'random'

# Preprocessing
preprocessing = 'none'
cutoff = (0.5 * np.float(intrinsic_dim)) ** 2
```

## Run Analysis

```python
%%capture
# Calculate Errors
error_normal, errors_spherical = run_error_analysis_inlier_noise(ambient_dim,
                   intrinsic_dim, inlier_points, outlier_points,
                   outlier_spreads, inlier_spreads, inlier_noise_spreads, repititions, sampling,
                   method, preprocessing = preprocessing, cutoff = cutoff)

errors_mean_normal = np.mean(error_normal, axis = 5)
errors_std_normal = np.std(error_normal, axis = 5)
errors_mean_spherical = np.mean(errors_spherical, axis = 5)
errors_std_spherical = np.std(errors_spherical, axis = 5)
```

## Plot Results

```python
y_index = 4
plot_error_inlier_noise(errors_mean_normal[:,:,:,:], errors_std_normal[:,:,:,:],
           errors_mean_spherical[:,:,:,:], errors_std_spherical[:,:,:,:],
           ambient_dim, intrinsic_dim, inlier_points, outlier_points[:],
           outlier_spreads, inlier_spreads, inlier_noise_spreads, repititions, sampling, method,
           y_index, preprocessing, cutoff)
```

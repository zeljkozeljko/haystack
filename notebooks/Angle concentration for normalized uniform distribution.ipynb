```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Functions
def uniform_sample_points(D, n_points):
    samples = np.random.uniform(low = -1.0, high = 1.0, size = (D, n_points))
    return samples

def gaussian_sample_points(D, n_points):
    samples = np.random.normal(size=(D, n_points))
    return samples

def normalize_samples(samples):
    normalized_samples = samples
    for j in range(samples.shape[1]):
        normalized_samples[:,j] = normalized_samples[:,j]/np.linalg.norm(normalized_samples[:,j])
    return normalized_samples
```

```python
# Often occurring vectors
def angles(normalized_vectors):
    D, n = normalized_vectors.shape
    angs = np.zeros(n)
    for i in range(n):
        angs[i] = angle(normalized_vectors[:,i])
    return angs
    
def angle(normalized_vector):
    D = normalized_vector.shape[0]
    signums = np.sign(normalized_vector).astype('float')
    reference_vector = signums * 1.0/np.sqrt(D)
    return np.arccos(np.dot(normalized_vector, reference_vector))
```

### Uniform Sampling in $[-1, 1]^D$

```python
# Run script
samples = uniform_sample_points(2, 100000)
normed_samples = normalize_samples(samples)
angs = angles(normed_samples)
fig = plt.figure()
bins = np.linspace(0,1.0,100)
plt.hist(angs, bins = bins)
plt.show()
```

### Gaussian Sampling $\mathcal{N}(0,\textrm{I}_D)$

```python
# Run script
samples2 = gaussian_sample_points(2, 100000)
normed_samples2 = normalize_samples(samples2)
angs = angles(normed_samples2)
fig = plt.figure()
bins = np.linspace(0,1.0,100)
plt.hist(angs, bins = bins)
plt.show()
```
